```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_csv("Agri.csv")
```

```python
df = df.drop([0])
```

```python
df1 = df.drop(df.iloc[:, 14:],axis = 1)
```

```python
soil = list(df['SOIL'].unique())
print(soil)
```

```python
df['SOIL_PH_LOW'].min()
```

```python
rice = df1[df1['CROPS'] == "rice"]
rice
```

```python
rice = rice.drop(['TYPE_OF _WATERSOURCE'], axis=1)
```

```python
for i in range(100):
    temp = {'CROPS': 'rice', 'TYPE_OF_CROP': 'cereals', 'SOIL': random.choice(soil)}
    rice = rice.append(temp, ignore_index = True)
```

```python
rice
```

```python
for i in rice.columns:
    if i not in ['CROPS', 'TYPE_OF_CROP', 'SOIL']:
        rice[i] = [ random.uniform(int(df[i].min()),int(df[i].max()))  for k in rice.index]
```

```python
rice
```

```python
crops = list(df['CROPS'].unique())
```

```python
def gen_value(df1, crop):
    for i in range(100):
        temp = {'CROPS': crop, 'TYPE_OF_CROP': 'cereals', 'SOIL': random.choice(soil)}
        df1 = df1.append(temp, ignore_index = True)
    for i in df1.columns:
        if i not in ['CROPS', 'TYPE_OF_CROP', 'SOIL']:
            df1[i] = [ random.uniform(int(df1[i].min()),int(df1[i].max()))  for k in df1.index]
    
    return df1
```

```python
df2 = df
```

```python
df2 = df2.drop(df2.iloc[:, 14:],axis = 1)
df2 = df2.drop(['TYPE_OF _WATERSOURCE'], axis=1)
```

```python
for i in crops:
    df2 = gen_value(df2, i)
```
